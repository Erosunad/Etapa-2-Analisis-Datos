```python
#Realizar grafica de dispersion

datos.plot.scatter(x="Sales", y="Units")
plt.show()
```

```python
#Regresi贸n lineal

regresion = linear_model.LinearRegression()
```

```python
#Se agregan los datos a un array

ventas = datos["Sales"].values.reshape((-1,1))
```

```python
#Se crea el modelo

modelo = regresion.fit(ventas, datos["Units"])

#Se imprime interseccion

print("Interseccion (b)", modelo.intercept_)

#Se imprime pendiente

print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[21],[24],[39],[44]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
datos.plot.scatter(x="Sales", y="Units", label= 'Informaci贸n Restaurantes')
plt.scatter(entrada, predicciones, color='green')
plt.plot(entrada, predicciones, color='red', label='Linea de Regresi贸n')
plt.xlabel('Sales')
plt.ylabel('Units')
plt.legend()
plt.show()
```
